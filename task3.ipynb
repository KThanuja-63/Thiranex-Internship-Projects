```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

from google.colab import files
uploaded = files.upload()

sales_data = pd.read_csv(next(iter(uploaded)))

print(sales_data.head())

sales_data['Date'] = pd.to_datetime(
    sales_data['Date'],
    format='%d-%m-%Y'
)

sales_data = sales_data.sort_values('Date')

sales_data['Month_Number'] = range(1, len(sales_data) + 1)

X = sales_data[['Month_Number']]
y = sales_data['Weekly_Sales']

prediction_model = LinearRegression()

prediction_model.fit(X, y)

future_month = [[len(sales_data) + 1]]

future_sales = prediction_model.predict(future_month)

print("Predicted Future Sales:", future_sales[0])

plt.figure(figsize=(8,5))

plt.scatter(X, y)

plt.plot(X, prediction_model.predict(X))

plt.xlabel("Time Progress")
plt.ylabel("Weekly Sales")

plt.title("Sales Forecast Analysis")

plt.show()
```
